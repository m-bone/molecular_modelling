```python
import os
from rdkit import Chem
from rdkit.Chem import Draw
from load_gdb import get_smiles_inchi

from IPython.display import display, Markdown

file_dir = "/home/mab-desk/Documents/atom_dbs/gdb9"
file_name = "dsgdb9nsd_017809.xyz"
os.chdir("/home/mab-desk/Documents/atom_dbs/gdb9")

smi, inchi = get_smiles_inchi(file_dir, file_name)

mol_smi_1 = Chem.MolFromSmiles(smi[0])
mol_smi_2 = Chem.MolFromSmiles(smi[1])
mol_inchi_1 = Chem.MolFromInchi(inchi[0])
mol_inchi_2 = Chem.MolFromInchi(inchi[1])
```

```python
def try_draw(mol, idx):
    order = {
        0: "Smiles 1",
        1: "Smiles 2",
        2: "InChI 1",
        3: "InChI 2"
    }

    if mol is None:
        display(Markdown(f"**{order[idx]} is Null**"))
    else:
        display(Markdown(f"**{order[idx]}**"))
        img = Draw.MolToImage(mol)
        display(img)

for idx, mol in enumerate([mol_smi_1, mol_smi_2, mol_inchi_1, mol_inchi_2]):
    try_draw(mol, idx)
```
